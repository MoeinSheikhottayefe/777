```python
import numpy as np
import matplotlib.pyplot as plt

# Angular frequency range from -2*pi to 2*pi
omega = np.linspace(-2*np.pi, 2*np.pi, 1000)

# Calculate complex number
z = np.exp(-1j * omega)

# Plot the unit circle
plt.figure(figsize=(6, 6))
plt.plot(np.real(z), np.imag(z), label=r'$e^{-j\omega}$')
plt.plot(np.cos(omega), np.sin(omega), linestyle='dashed', color='gray', label='Unit Circle')
plt.xlabel('Real Part')
plt.ylabel('Imaginary Part')
plt.title('Complex Number $e^{-j\omega}$ for $-2\pi \leq \omega \leq 2\pi$')
plt.axis('equal')
plt.grid(True)
plt.legend()
plt.show()
```
